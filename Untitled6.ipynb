```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')  # or 'titanic.csv'
df.head()
```

```python
df.info()  # Check data types and nulls
df.describe()  # Summary statistics
df.isnull().sum()  # Count missing values
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)  # too many missing values
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode 'Sex'
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])  # Male = 1, Female = 0

# One-hot encode 'Embarked'
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
# Boxplot for outliers
sns.boxplot(df['Fare'])
plt.title("Fare Outliers")
plt.show()

# Remove extreme outliers if necessary
df = df[df['Fare'] < 300]  # example threshold
m
```
